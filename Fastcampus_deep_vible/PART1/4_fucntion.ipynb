#1.Fucntion
승차권 예제를 가지고 function에 대해서 실습해 보자 

```python
# 저번 시간 반복
stationfares = [
                {'station': "청량리", "fare": 1500},
                {'station': "성북", "fare": 1800},
                {'station': "의정부", "fare": 2000},
                {'station': "소요산", "fare":2500}
    ]
```

```python
stations = []
fares = []
```

```python
for stationfare in stationfares:
    stations.append(stationfare['station'])
    fares.append(stationfare['fare'])
print(stations)
print(fares)
```

##내장함수

```python
# 길이 출력
len(stationfares)
```

```python
# 최대값 출력
max(fares)
```

```python
# 최솟값 출력
min(fares)
```

```python
# list의 총합 출력
sum(fares)
```

```python
# 인덱스 만들어주기: enumerate (중요!)
# 아래 3개 for문의 차이점 기억하기 
for station in stations:
    print(station)

for station in enumerate(stations): 
    print(station)

for idx, station in enumerate(stations):
    print(idx, station)
```

##사용자 정의 함수
getfare() 함수 정의하기 

```python
# 이런 식으로 함수를 만드는 이유는 '재활용성'이 좋기 때문이다. 
# 또한 식이 잘못되었을 때 식을 할당한 식들을 모두 바꾸는 것이 아닌
# 함수 안의 수식만 변경하면 되므로 수정이 편해진다. 
def getFare(station):
    for stationfare in stationfares:
        if stationfare['station'] == station:
            return stationfare['fare']
fare = getFare('청량리')
print(fare)
```

```python
# station을 가지고와서 getFare를 활용, 역마다 fare 출력하기 
for station in stations:
    print(station, getFare(station))
```
