## 1.Function
우선 Function에 대한 복습을 하면서 진행해보자

```python
stationfares = [
                {"station": "청량리", "fare": 1500},
                {"station": "성북", "fare": 1800},
                {"station": "의정부", "fare": 2000},
                {"station": "소요산", "fare": 2500}
]
```

```python
stations = []
fares = []
for stationfare in stationfares:
    stations.append(stationfare["station"])
    fares.append(stationfare["fare"])
print(stations)
print(fares)
```

```python
len(stationfares)
```

```python
max(fares)
```

```python
min(fares)
```

```python
sum(fares)
```

```python
for idx, station in enumerate(stations):
    print(idx, station)
```

```python
sum = lambda a, b : a+b
sum(2,3)
```

```python
def getFare(station):
    for stationfare in stationfares:
        if stationfare['station'] == station:
            return stationfare['fare']
fare = getFare("청량리")
print(fare)
```

```python
for station in stations:
    print(station,getFare(station))
```
