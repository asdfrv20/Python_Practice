#1. OpenAPI

고속도로 공공데이터 포털의 "실시간 영업시간 통행시간"

Reauest URL: http://data.ex.co.kr/openapi/trtm/realUnitTrtm

고속도로 공공데이터 포털의 '실시간 영업소간 통행시간' url: http://data.ex.co.kr/openapi/basicinfo/openApiInfoM?apiId=0147&serviceType=&keyWord=&searchDayFrom=2014.12.01&searchDayTo=2021.07.21&CATEGORY=TR&GROUP_TR=TIME_TCS

```python
import requests
```

```python
# 요청변수 입력 
# '실시간 영업소간 통행시간' page를 보고 형식에 맞춰 넣어주기 
key = '8738210265'
type = 'json'
StartUnitCode = '101'
EndUnitCode = '103'

# '실시간 영업소간 통행시간'page 하단에 '예제 실행하기'에 key, type, iStartUnitCode, iEndUnitCode를 입력 후 URL 보기를 하면 나오는 형식에 따라 구성해 주어야 한다.
URL = 'http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key='
url = URL + key + '&type=json&iStartUnitCode=' + StartUnitCode + '&EndUnitCode=' + EndUnitCode
```

```python
response = requests.get(url)    #  위에서 만든 url을 get방식으로 던져주어 response에 결과값을 할당해주기 
```

```python
print(response)
```

```python
# 가져온 response를 "json 형식"으로 표현하기 & 내용 확인하기 
# 우리가 필요한건 json 안에 list안에 들어있는 데이터(realUnitTrtmV0)
json = response.json()
json
```

```python
# 차들에 대한 list만 가져오기 
cars = json['realUnitTrtmVO']
cars
```

```python
# (for문을 돌려서) car에서 날짜, 시간, 차 타입만 가져오기
# 날짜(stdDate), 시간(stdTime), 차 타입(tcsCarTypeDivName)
records = []
for car in cars:
    dic = {}
    dic['date'] = car['stdDate']
    dic['time'] = car['stdTime']
    dic['type'] = car['tcsCarTypeDivName']
    records.append(dic)
records
```
