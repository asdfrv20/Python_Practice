```python
# Form을 적용한 경우
# #@param {type: "string"}을 추가해 줌으로써 왼쪽에 보이는 form에 입력 시 코드도 변경이 됨.
name = 'Tom' #@param  {type:"string"}
names.append(name)
print(names)
```

#2. Dictionary
승차권 판매기 실습 

```python
stationfares = []
```

```python
dic = {}
station = input('Enter station : ')
fare = int(input('Enter fare: '))
dic['station'] = station
dic['fare'] = fare
stationfares.append(dic)
print(dic)
print(stationfares)
```

```python
# Form을 활용한 입력형태
dic = {}
station = '\uC131\uBD81' #@param {type:"string"}
fare = '2000' #@param {type:"string"}
dic['station'] = station
dic['fare'] = fare
stationfares.append(dic)
print(dic)
print(stationfares)
```
