## for 문

```python
# 구구단 실습
num = int(input('Multimultication table of : '))
```

```python
for i in range(1,10):
    print(num, 'x', i, '=', num*i)
```

```python
# for 중첩 반복문
for i in range(2,10):
    for j in range(1,10):
        print("%2d X %2d = %5d" %(i, j, i*j))
    print('----------------')
```

## while 문

```python
num = int(input('Multiplicaiton table of : '))
```

```python
i = 1
while i<10:
    print("%2d * %2d = %5d" %(num, i, num*i))
    i += 1 
```

```python
i = 2
# j = 1  # 내가 j를 넣은 부분
while i<10:
    print("[%d단]" %i)
    j = 1  # lecture에서 j를 초기화 시켜준 부분 
    while j<10:
        print("%2d * %2d = %5d" %(i, j, j*i))
        j += 1
    print("-----------------------")
    i += 1
    # j = 1  # 내가 j를 넣은 부분 
```

# 승차권 발매 변형 예제
다차원의 군집데이터를 다시 1차원으로 분화해 나가기 (for문 활용)

```python
stationfares = [
           {'station': "청량리", "fare": 1500},
           {'station': "성북", "fare": 1800},
           {'station': "의정부", "fare": 2000},
           {'station': "소요산", "fare": 2500}
        ]
```

```python
stations = []
fares = []
for stationfare in stationfares:
    stations.append(stationfare["station"])
    fares.append(stationfare["fare"])
```

```python
print(stations)
print(fares)
```
