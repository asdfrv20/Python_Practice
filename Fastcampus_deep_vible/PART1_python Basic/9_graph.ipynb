#2.Class
요금을 넣으면 역에 맞게 요금을 차감하는 계산기를 만들어보자

##class 생성하기
Comments, Products, SubwayTickent

```python
class Comments:
    title               = "#### %s 승차권 판매기 입니다. ####"                          # title: 승차권 판매기 시작 멘트
    product_description = "%s: %s(%s원)"                                                # product_description: 도착지, 금액 출력
    insert_fare         = "\n 요금을 넣어주세요. : "                                    # insert_fare: 요금 입력 안내 문구
    insufficient_fare   = "%s 요금이 부족합니다. 거스름돈은 %s 원 입니다."              # insufficient_fare: 요금이 부족할 경우, 출력하는 멘트
    select_station      = "원하시는 역번호를 입력하세요."                               # select_station: 도착지(역) 입력 멘트
    select_error        = "잘못 입력하셨습니다."                                        # select_error: 선택 오류 시 출력
    finish_sale         = "선택하신 %s 입니다. 거스름돈은 %s원 입니다. \n 감사합니다."  # finish_sale: 거래 후. 승차권과 거스름돈 반환 메세지
    terminate_sale      = "승차권 판매를 종료합니다."                                  # terminate_sale: 승차권 판매 종료 멘트
```

```python
class Products:                 # ~~ 승차권의 정보가 들어갈 class
    productNames = []           # 도착지(역) 이름
    productValues = []          # 도착지 별 가격
```

```python
# '지하철' 승차권의 설계도 라고 생각하면 됨. 
class SubwayTicket(Products):
    _data = stationfares        # _data: 저장될 data(기존의 역-금액 data를 사용할 것임. )
    _name = "지하철"            # _name: 어떤 것의 승차권 판매기 인가? (지하철 승차권 판매기이기 때문에 _name = "지하철"로 설정)
    _status = True              # _status: 계속 반복을 할 것인가? (True: 반복, False: 반복X)

    def __init__(self):                   # 생성자
        print(Comments.title %self._name)   # '_name' + Comments.title  출력
    """
    [승차권 판매기 동작 순서]
    0. set_products 실행: _data(stationfares)를 productNames(station), productValues(fare)로 나누어 저장한다.
    1. run: 요금을 입력받고, 입력 값에 이상이 없다면 selectStation 매서드를 호출한다.  
    2. selectStation:(input: money) 도착역들의 리스트를 출력하고 사용자에게 도착역을 입력받는 매서드. 입력이 이상이 없다면 payment 매서드를 호출한다. 
    3. payment: (input: money, idx) 입력 금액과 필요요금을 비교하고, 승차권과 잔금을 반환한다. (+run의 while문 반복)
    """
    def set_products(self):               # set_products:  _data(stationfares)를 가지고와서 Products.productNames(station 값 저장)와 Products.productValues(fare 값 저장)로 나누는 매서드 
        Products.productNames = []          # 해당 매서드를 부를 때 마다. Products.productsNames와 Products.productValues를 초기화 시킨다(일다 비우고 시작)
        Products.productValues = []
        for stationfare in self._data:      # stationfares에 들어있는 데이터를 각각 productNames에는 station을 productValus에는 fare를 저장한다.  
            Products.productNames.append(stationfare["station"])
            Products.productValues.append(stationfare["fare"])

    def run(self):              # run: 승차권을 판매하는 핵심 메서드. 요금을 입력받고 이상이 없다면, selectStation 메서드를 호출한다.  
        self.set_products()     # 0. set_products(): station과 fare에 대한 정보(productNames, productValeus)를 setting 시켜준다. 
        while self._status:     # self._status의 상태에 따라 while문이 반복될지 결정된다. 
            try:                                                    # 1. 요금 입력 문구를 출력한다.  
                inputMoney = int(input(Comments.insert_fare))       #### 요금 입력 안내 문구를 출력하고, inputMoney를 받는다(Comments.insert_fare) 
                if inputMoney < 0:
                    raise ValueError
            except ValueError:                                      # 2-1) ValueError: 값 입력이 이상한 경우 발생하는 오류
                print(Comments.select_error)                        ####               '잘못 입력하셨습니다.'를 출력한다.  
            else:                                                   # 2-2) 오류가 아닌 경우,
                self.selectStation(inputMoney)                      ####               selectSattion(inputMoney) 매서드를 실행시킨다.

    def selectStation(self, money):          # selectStation: (input: money) 원하는 역을 선택하고 그 정보를 출력하는 매서드
        print(Comments.select_station)       # 0. '원하시는 역 번호를 입력하세요' 메세지 출력(Comments.seletct_station) + 모든 역들의 정보 출력(for문)
        for idx, name in enumerate(Products.productNames):                  # productNames(station에 대한 정보) 데이터를 index화 시켜서 반복 실행>> productNames와 productValues의 인덱스가 같음을 이용
            fare = Products.productValues[idx]                              # fare에 금액(productValues) 입력, idx를 활용하여 편하게 설정
            print(Comments.product_description %(str(idx), name, fare))     # station의 전체 정보 출력: "역번호[idx]: 도착역이름[station](요금[fare])" (Comments.select_station)
        inputStation = int(input(Comments.select_station))                  # inputStation: 사용자로부터 선택받은 도착역
        self.payment(money, inputStation)                                   # payment(money, inputStation) 매서드 호출

    def payment(self, money, idx):          # payment: (input: money, idx) 입력한 돈과 승차권의 값을 비교하고, 승차권을 출력하도록 하는 매서드
        name = Products.productNames[idx]   # name과 fare를 가져온다. 이 때 우리는 idx를 가져왔기 때문에 쉽게 가져올 수 있음(productNames와 productValues에서 가져오기)
        fare = Products.productValues[idx]
        if money >= fare:                                           # 입력받은 money >= 필요요금 일 때: 잔금(balance)과 승차권을 반환한다.   
            balance = money - fare                                  # 선택한 금액과 거스름돈을 출력하는 메세지[선택하신 %s 입니다. 거스름돈은 %s원 입니다. \n 감사합니다.](Comment.finish_sale(도착역, balance: 남은 돈))
            print(Comments.finish_sale %(name, str(balance)))   
        else:                                                       # 입력 money < 필요요금 일때: 요금 부족 메세지와 함께 입력 금액을 반환한다.
            print(Comments.insufficient_fare %(name, str(money)))   # 금액 불충분 메세지[%s 요금이 부족합니다. 거스름돈은 %s 원 입니다.](Comment.insufficient_fare(도착역, 요금))를 출력 

## 여기까지가 말 그대로 설계도!!(class 작성은 끝!!!)

```

#3.Inheritance
상속, 위에서 만든 class를 사용하여 KTX의 승차권 판매기를 만들어보자

```python
railstationfares = [
        {"station": "천안", "fare": 15000},
        {"station": "대전", "fare": 20000},
        {"station": "동대구", "fare": 37000},
        {"station": "부산", "fare": 53000}
]
```
